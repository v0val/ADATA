2.1 Изучите распределение домов от наличия вида на набережную. Постройте график

```python
data = df.groupby('price')['waterfront'].sum().reset_index()
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.scatter(x=data['waterfront'], y=data['price'])

plt.title('График разброса значений')
plt.xlabel('Стоимость')
plt.ylabel('Вид на набережную');
```

2.2 Изучите распределение этажей домов

```python
plt.figure(figsize=(6, 4))
plt.hist(df['floors'], bins = 15)
plt.title('Распределение количества этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество квартир');
```

2.2 Изучите распределение состояния домов
